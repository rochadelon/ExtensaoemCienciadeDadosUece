```
# Fixing the NameError by ensuring the 'filtrar_coluna' function is defined and updating the code.

import pandas as pd
import google.generativeai as genai

# Define the 'filtrar_coluna' function
def filtrar_coluna(df, coluna, texto, tipo='contem'):
    """
    Filtra a coluna especificada em um DataFrame com base no texto e no tipo definidos.
    
    :param df: DataFrame que será filtrado.
    :param coluna: O nome da coluna a ser filtrada.
    :param texto: O texto usado para filtrar os dados.
    :param tipo: O tipo do filtro ('contem', 'comeca_com', 'termina_com').
    :return: DataFrame filtrado.
    """
    if tipo == 'contem':
        return df[df[coluna].str.contains(texto, case=False, na=False)]
    elif tipo == 'comeca_com':
        return df[df[coluna].str.startswith(texto, na=False)]
    elif tipo == 'termina_com':
        return df[df[coluna].str.endswith(texto, na=False)]
    else:
        raise ValueError("Tipo de filtro não suportado. Use 'contem', 'comeca_com' ou 'termina_com'.")

# Configure the Gemini API
genai.configure(api_key='AIzaSyAcQicBEj8s2NvvCBORlQCgPUrPchR-vCQ')

# Define the Gemini analysis function
def analisar_com_gemini(dados_filtrados):
    """Envia dados filtrados para análise do Gemini"""
    # Converte os dados para texto
    texto_dados = dados_filtrados.to_string()

    model = genai.GenerativeModel('gemini-pro')

    prompt = f"""
    Analise estes dados sobre 'words scope of paints and nanocoatings':

    {texto_dados}

    Forneça insights sobre os padrões encontrados.
    """

    response = model.generate_content(prompt)
    return response.text

# Example usage (assuming 'df' DataFrame is already loaded and properly defined):
texto_filtro = "words scope of paints and nanocoatings"

# Example: Creating a dummy DataFrame to simulate input
data = {'titulo': ["Words scope of paints and nanocoatings improve.",
                   "Exploring scope of other industries.",
                   "Nanocoatings in modern technology."]}
df = pd.DataFrame(data)

# Filter the DataFrame
dados_filtrados = filtrar_coluna(df, 'titulo', texto_filtro, tipo='contem')

print("Dados filtrados:")
print(dados_filtrados)

# Send filtered data for Gemini analysis
analise = analisar_com_gemini(dados_filtrados)
print("Análise do Gemini:")
print(analise)
```
