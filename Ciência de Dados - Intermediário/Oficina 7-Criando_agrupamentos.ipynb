## K-means: Vantagens e Limitações

O algortimo escolhido foi o k-mean segundo as características dos dados, contínuo, sem presença de outliers e com o número de clusters, k=3, já conhecido. 

As vantagens do k-means é sua simplicidade e eficiência computacional mesmo em situações onde há um enorme quantidade de dados, entretando é muito sensível a outliers e não lida bem com dados não linates.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D
```

```
np.random.seed(0)
n_samples = 300

data = {

    'valor_medio_gasto': np.random.randn(n_samples) * 20 + 200,

    'frequencia_compras': np.random.randn(n_samples) * 5 + 15,

    'categorias_produtos': np.random.randn(n_samples) * 2 + 8

}

df = pd.DataFrame(data)

df.head()
```

```
X = df[['valor_medio_gasto', 'frequencia_compras']]
```

```
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
df['cluster'] = kmeans.labels_
```

```
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['cluster'], s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')

plt.title('Segmentação de Clientes por Comportamento de Compra')
plt.xlabel('Valor Médio Gasto')
plt.ylabel('Frequência de Compras')
plt.show()
```

## Insights e Análise dos Clusters

Foram gerados três clusteres a partir do Valor Médio Gasto e a Frequência de Compras. Esses grupos permite encontrar 3 grupos estratégicos. 

Cluster 1 (verde - Ocasionais): representa os clientes ocasionais, gastam pouco e com baixa frequência. Esses clientes podem ser atraídos por promoções que visem aumentar a frequência das compras. 

Cluster 2 (roxo - Regulares): Esses são os clientes regulares, gastam moderadamente com alta frequência. Pode ser eficaz criar programas de fidelização com promoções e descontos progressivos a fim de aumentar o valor de cada compra.

Cluster 3 (amarelo - Especiais): Representa os clientes especiais que compram alto valores com uma ótima frequência.Aqui o ideal é criar estratégias retenham esses cliente e o insentivem a comprar ainda mais frequentemente sem reduzir o valor médio gasto, como clube de vantagens exclusivas. atendimento personalizado, compras programadas e brindes.
