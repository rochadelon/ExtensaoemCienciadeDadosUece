## Correlação de Pearson

Como um conjunto de dados não é muito grande, você pode calcular a coeficiente de correlação padrão (Pearson) entre cada atributo da base de dados.

O coeficiente varia de -1 a 1 e pode ser interpretado da seguinte forma:
Potência de 1: forte correlação positiva
Potência de -1: forte correlação negativa
Potência de 0: Nenhuma correlação linear entre os dados
